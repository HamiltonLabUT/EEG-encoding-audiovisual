```python
def errorbars(x, data, color, alpha=0.3):
	'''
	x = your x axis (vector of x coordinates)
	data = your data value (y)
	'''
	sem_above = data.mean(0) - data.std(0)/np.sqrt(data.shape[0])
	sem_below = data.mean(0) + data.std(0)/np.sqrt(data.shape[0])
	plt.plot(x, data.mean(0), color)
	plt.fill_between(x, sem_below, sem_above, color=color, alpha=alpha)
```

```python
#plot grand avg. response
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
tmin=-0.3
tmax=0.5
colors = ['green', 'red']
t = np.linspace(tmin,tmax,num=np.array(congruent_av).shape[1])
plt.figure()

#audiovisual
errorbars(t, np.array(congruent_av), colors[0],alpha=0.4)
errorbars(t, np.array(incongruent_av), colors[1], alpha=0.4)

#visual only
errorbars(t, np.array(congruent_v), colors[0], alpha=0.2)
errorbars(t, np.array(incongruent_v), colors[1], alpha=0.2)

plt.axvline(0,color='k')
plt.axhline(0,color='k')
plt.legend(['congruent','incongruent'])
plt.xlabel('Time [s]')
plt.ylabel('Amplitude [uV]')
#plt.savefig(f'{save_fig}/python_figs/{elec_type}_grandAvg_mouthing-ERPs_{condition}.pdf')
```
