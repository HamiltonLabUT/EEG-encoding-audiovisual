## Step 2 of Audio Preprocessing

add the visual only offset times to the event file generated in step 1 of audio preprocessing

```python
import numpy as np
import pandas as pd
```

```python
Subject = 'MT0028' #change to subject you want to load 
User = 'alyssa' #change for your path 
```

```python
#Load the event file for the subject 
mt_event = pd.read_csv(f'/Users/{User}/Library/CloudStorage/Box-Box/MovieTrailersTask/Data/EEG/Participants/{Subject}/audio/{Subject}_B1_MovieTrailers_events.csv', sep=",", header = 0)
mt_event = mt_event.dropna(axis='columns') #drop columns which appear as Nan
mt_event.columns = ["name", "path_name","event_id", "onset_time", "offset_time", "match_filter_corr"] #let's add in names of the columns so it's easier to read into once we loop through

```

```python
mt_event
```

```python
#loop through the files:

# Make new event to append to 
new_mt_event = {'name': [], 'path_name': [], 'event_id': [],'onset_time': [], 'offset_time': [], 'match_filter_corr' : [] }

for row_number, row in enumerate(mt_event.values):
    mt_name = row[1]
    # If it's a visual only file, find the trailer *with* audio
    # and use it to get the duration, and modify the duration
    # based on onset of the visual only + whatever the duration
    # is calculated to be
    if '_visualonly_notif' in mt_name:
        #print(row_number, row)
        mt_name_orig = mt_name.replace('_visualonly_notif', '')
        print(mt_name, mt_name_orig)
        
        match_rows = np.where(mt_event['path_name'] == mt_name_orig)[0][0]
        print(match_rows)
        duration = mt_event.iloc[match_rows]['offset_time'] - mt_event.iloc[match_rows]['onset_time'] 
        print(duration)
        
        onset = mt_event.iloc[row_number]['onset_time']
        offset = onset + duration
        event_id = mt_event.iloc[row_number]['event_id']
        wavname = mt_event.iloc[row_number]['path_name']
        trial_name = mt_event.iloc[row_number]['name']
        match_corr = mt_event.iloc[row_number]['match_filter_corr']
        new_mt_event['onset_time'].append(onset)
        new_mt_event['offset_time'].append(offset)
        new_mt_event['event_id'].append(event_id)
        new_mt_event['path_name'].append(wavname)
        new_mt_event['name'].append(trial_name)
        new_mt_event['match_filter_corr'].append(match_corr)
        
    else:
        # If it's a normal movie trailer with audio, don't change anything,
        # just make a copy
        onset = mt_event.iloc[row_number]['onset_time']
        offset = mt_event.iloc[row_number]['offset_time']
        event_id = mt_event.iloc[row_number]['event_id']
        wavname = mt_event.iloc[row_number]['path_name']
        trial_name = mt_event.iloc[row_number]['name']
        match_corr = mt_event.iloc[row_number]['match_filter_corr']
        new_mt_event['onset_time'].append(onset)
        new_mt_event['offset_time'].append(offset)
        new_mt_event['event_id'].append(event_id)
        new_mt_event['path_name'].append(wavname)
        new_mt_event['name'].append(trial_name)
        new_mt_event['match_filter_corr'].append(match_corr)
    
new_mt_df = pd.DataFrame.from_dict(data=new_mt_event)
```

```python
#Save new event file as a csv 
new_mt_df.to_csv(f'/Users/{User}/Library/CloudStorage/Box-Box/MovieTrailersTask/Data/EEG/Participants/{Subject}/audio/{Subject}_B1_MovieTrailers_events.csv')
```
